```python
 
####################
## Tarea en clase ##
####################

# Cargar conjunto de datos mediante pandas

import pandas as pd
df_original = pd.read_excel('Consumo Gasolina Autos Ene 2018.xlsx')
df_original
```

```python
# Especificar el tipo de datos de cada columna.

df_original.dtypes
```

```python
# Si es necesario, asignar una columna como índice.

df_1 = df_original.reset_index()
df_1
```

```python
#############
## Tarea 2 ##
#############

# Crear, modificar o eliminar columnas


# Se genera lista booleana hibrido con los vehículos hibridos de acuerdo a su Versión o Submarca
hibrido1 = ['HIBRIDO' in x or 'HYBRID' in x or 'HÍBRIDO' in x for x in df_1['Versión'] ]
hibrido2 = ['HIBRIDO' in x for x in df_1['Submarca']]
hibrido = []
for i in range(len(hibrido1)):
    if hibrido1[i] or hibrido2[i]:
        hibrido.append(1)
    else:
        hibrido.append(0)

# Se crea columna 'Hibrido' con 1 si el vehículo es híbrido y 0 si no es híbrido.
df_1['Hibrido'] = hibrido

# Se remueve la columna index que se creó ya que no es necesaria.
df_1 = df_1.drop('index', axis=1)

df_1.head(3)
```

```python
# Se revisa si hay valores faltantes en la tabla.
df_1.isnull().sum()
```

```python
# Se encuentran 16 valores nulos en Transmisión ('Trans.')
# Se procede a eliminar esos 16 registros.

df_2 = df_1.dropna()
df_2
```

```python
# La variable 'Calificación Contam. Aire' es de tipo object, cuando se esperaba que fuera int64.
# Se revisa que tiene 1,233 registros con valor ? lo cual se debe tratar como un valor nulo.
df_2[df_2['Calificación Contam. Aire']=='?']['Calificación Contam. Aire'].count()
```

```python
# Procedemos a remover esta variable 'Calificación Contam. Aire' de nuestro dataframe.

df_3 = df_2.drop('Calificación Contam. Aire', axis=1)
df_3.head(3)
```

```python
# Tratamiento de variables categoricas

df_var_cat = df_3.select_dtypes(include='object')

for col in df_var_cat.columns:
    print(col, ': ', len(df_2[col].unique()), ' labels')
```

```python
# Dado que hay muchas categorías de Versión que no permite un análisis al casi tener cada vehículo su propia Versión, se procederá a remover este campo.

# De igual forma como hay muchas categorías de Submarca y estas se encuentran agrupadas por Marca, se procederá a remover este campo.

df_var_cat2 = df_var_cat.drop(['Submarca', 'Versión'], axis=1)
df_var_cat2.head(3)
```

```python
# Aplicamos One Hot Encoding a las variables 'Comb.' y 'Cateogría'.

df_var_cat3_1 = df_var_cat2.drop(['Marca', 'Trans.'], axis=1)
df_var_cat3_1 = pd.get_dummies(df_var_cat3_1, drop_first=True)
df_var_cat3_1 = df_var_cat3_1.astype('int64')
df_var_cat3_1
```

Para las variables 'Marca' y 'Trans.' aplicaremos One Hot Encoding pero seleccionando el top 15 y top 10 en frecuencia, respectivamente,
ya que con esto se abarca al menos el 75% de los registros y el resto se agrupan en una sola categoría.



```python
df_var_cat2['Marca'].value_counts().sort_values(ascending=False).head(15).sum() / 4601
```

```python
df_var_cat2['Trans.'].value_counts().sort_values(ascending=False).head(10).sum() / 4601
```

```python
# Top 15 'Marca'
df_var_cat3_2 = df_var_cat2.drop(['Comb.', 'Categoría'], axis=1)
top_15 = [x for x in df_var_cat3_2['Marca'].value_counts().sort_values(ascending=False).head(15).index]
top_15
```

```python
# Top 10 'Trans.'
top_10 = [x for x in df_var_cat3_2['Trans.'].value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
# Ahora aplicamos el One Hot Encoding a estos tops.
import numpy as np

for label in top_15:
    df_var_cat3_2['Marca_'+label] = np.where(df_var_cat3_2['Marca']==label, 1, 0)
df_var_cat3_2 = df_var_cat3_2.drop('Marca', axis=1)

for label in top_10:
    df_var_cat3_2['Trans._'+label] = np.where(df_var_cat3_2['Trans.']==label, 1, 0)
df_var_cat3_2 = df_var_cat3_2.drop('Trans.', axis=1)

df_var_cat3_2.head(3)
```

```python
# Concatenamos todas las variables en un solo df.

# df_3.reset_index(inplace=True)
# df_var_cat3_1.reset_index(inplace=True)
# df_var_cat3_2.reset_index(inplace=True)


df_var_cat4 = df_var_cat3_1.join(df_var_cat3_2)
df_4 = df_3.join(df_var_cat4)
df_final = df_4.drop(columns=['Marca', 'Submarca', 'Versión', 'Trans.', 'Comb.', 'Categoría'], index=1)
df_final.columns
```

```python
# Guardamos algunos de los df principales para uso posterior.
df_3.to_csv('datos.csv', index=False)
df_final.to_csv('datos_proc.csv', index=False)
df_4.to_csv('datos_completo.csv', index=False)
```

```python
######################################################
# Procesamiento Alternativo de Variables Categóricas #
######################################################

# Debido a la dificultad para revisar/analizar en posteriores tareas la gran cantidad de variables dummies que se generaron
# en este apartado se opta mejor por redefinir las variables categóricas mediante el método Mean Encoding.

variables_cat = ['Marca', 'Submarca', 'Versión', 'Trans.', 'Comb.', 'Categoría']

for var in variables_cat:
    # 1.
    Encoded_Mean_Dict = df_3.groupby([var])['CO2(g/km)'].mean().to_dict()
    # 2.
    df_3[var + '_Mean_Encoded'] = df_3[var].map(Encoded_Mean_Dict)

# 3.
df_5 = df_3.drop(columns=variables_cat, axis = 1)
df_5.head()
```
